```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

print("All libraries loaded successfully!")
```

```python
df = pd.read_csv('raw_data/ethiopia.csv')
df.head()
```

```python
# Numeric summary
df.describe()

# Count missing values
missing = df.isna().sum()
missing_percent = (missing / len(df)) * 100
missing_percent[missing_percent > 5]  # Columns with >5% nulls
```

```python
cols = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']
z_scores = np.abs(stats.zscore(df[cols].dropna()))
outliers = (z_scores > 3).any(axis=1)
df['outlier_flag'] = outliers
```

```python
for col in cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
df_clean = df[~df['outlier_flag']]
```

```python
import os

# Create the folder if it doesn't exist
os.makedirs("data", exist_ok=True)

# Now save your cleaned CSV
df_clean.to_csv("data/ethiopia_clean.csv", index=False)
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=df_clean, x='Timestamp', y='GHI', label='GHI')
sns.lineplot(data=df_clean, x='Timestamp', y='DNI', label='DNI')
sns.lineplot(data=df_clean, x='Timestamp', y='DHI', label='DHI')
plt.show()
```

```python
df_clean['month'] = df_clean['Timestamp'].dt.month
monthly_avg = df_clean.groupby('month')['GHI'].mean()
monthly_avg.plot(kind='bar')
```

```python
sns.boxplot(x='outlier_flag', y='ModA', data=df)
sns.boxplot(x='outlier_flag', y='ModB', data=df)
```
